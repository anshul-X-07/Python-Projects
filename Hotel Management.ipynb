```python
# Just a decorative gesture
def welcome_message():
    print("*"*41)
    print("*\tHotel Management System\t\t*")
    print("*"*41)
```

### Hotel Class

```python
# hotel class
# you can access details like avaliable rooms, reserved rooms, searching rooms
class hotel:
        # to initialize the objects (like constructor)
        def __init__(self, availRooms, reservedRooms, searchRooms):
            self.availRooms = availRooms
            self.reservedRooms = reservedRooms
            self.searchRooms = searchRooms
            
availRooms = []
# total 100 available rooms are there in hotel 1st room starts with 1 and as follows to 100
for i in range(1, 101):
    availRooms.append(i)

# empty list to hold reserved rooms 
reservedRooms = []
# empty dict to hold searching rooms 
searchRooms = {}

# below line means "Object h is created of class hotel"
h = hotel(availRooms,reservedRooms,searchRooms)
```

### Customer Class

```python
# customer class

# like database to hold customer info
data = {}

class customer:
    # to initialize the objects (like constructor)
    def __init__(self, name, phNum, rooms, stayDuration, bookedRoom):
        self.name = name
        self.phNum = phNum
        self.rooms = rooms
        self.stayDuration = stayDuration
        self.bookedRoom = bookedRoom
    
    # method to input_data in data, name as key   
    def input_data(self):
        key = self.name
        data[key] = [self.phNum, self.rooms, self.stayDuration,self.bookedRoom]
    
```

### Booking Menu

```python
def booking():
    while True:
        print("*** Booking Menu ***")
        print(" 1. No. of rooms available ")
        print(" 2. No. of rooms reserved ")
        print(" 3. Make new booking ")
        print(" 4. Go to Home menu ")
        print()
        
        choice = int(input("Enter your choice :: "))
        print()
        
        if choice == 1:
            # len function returns the length of any datatype
            print("----------------------------------------------------")
            print("*** No. of rooms available are : ",len(h.availRooms)," ***")
            print("----------------------------------------------------")
            print()
        elif choice == 2:
            print("----------------------------------------------------")
            print("*** No. of rooms reserved are : ",len(h.reservedRooms)," ***")
            print("----------------------------------------------------")
            print()
        elif choice == 3:
            bookRoom()
            print()
        elif choice == 4:
            # goes back to the home menu 
            #(if you call here start again , after break of parent loop, this present function will still be on loop)
            return
            print()
        else:
            print("xxxxxx Invalid Input xxxxxx")
            print()
        
```

```python
# booking for new room
def bookRoom():
    print("<====> Enter new room details <====>")
    name = input("Customer Name : ")
    phNum = input("Contact No. : ")
    rooms = int(input("No. of rooms to book : "))
    stayDuration = int(input("Duration stayed (days) in room : "))
    
    bookedRooms = []
    for i in range(0,rooms):
        # available room will be booked and pops out from available rooms list
        roomNo = h.availRooms.pop(0)
        # hold the booked rooms number
        bookedRooms.append(roomNo)
        # this will help to find which customer booked which room no.
        h.searchRooms[roomNo] = name
        # finally all booked room will be appended to reserved rooms list to keep track of reserved rooms
        h.reservedRooms.append(roomNo)
        
    cust = customer(name, phNum, rooms, stayDuration,bookedRooms)
    cust.input_data()
    
    print()
    print("----------------------------------------------------")
    print("###       Successfully Booked A New Room         ###")
    print("----------------------------------------------------")
    print()
    print(":: Recent Booking Preview :: ")
    print(f"Name : {name} \nContact No. : {data[name][0]} \nNo. of rooms : {data[name][1]} \nDuration of Staying : {data[name][2]} \nRooms Booked : {data[name][3]} \n")
    print("----------------------------------------------------")
    print()
    
```

### Enquiry Menu

```python
def enquiry():
    while True:
        print("*** Enquiry Menu ***")
        print(" 1. Search by Customer Name ")
        print(" 2. Search by Rooms Number ")
        print(" 3. Check for Room Availability ")
        print(" 4. Go to Home Menu ")
        
        choice = int(input("Enter your choice :: "))
        print()
        
        if(choice == 1):
            searchByCustomerName()
            print()
        elif(choice == 2):
            searchByRoomNo()
            print()
        elif(choice == 3):
            checkAvailability()
            print()
        elif(choice == 4):
            # goes back to the home menu 
            #(if you call here start again , after break of parent loop, this present function will still be on loop)
            return
            print()
        else:
            print("xxxxxx Invalid Input xxxxxx")
            print()
            
```

```python
# search the database by customer name 
def searchByCustomerName():
    name=input("Enter Customer Name : ")
    print()
    
    # perviously in database name is set as key 
    if name in data.keys():
        print("==== Record Found ====")
        print(f"Name : {name} \nContact No. : {data[name][0]} \nNo. of rooms : {data[name][1]} \nDuration of Staying : {data[name][2]} \nRooms Booked : {data[name][3]} \n")
        print("----------------------------------------------------")
        print()
    else:
        print("----------------------------------------------------")
        print("xxxx             Record Not Found               xxxx")
        print("----------------------------------------------------")
        print()
        return
```

```python
# search the customer details by room number 
def searchByRoomNo():
    roomNo=int(input("Enter the Room number to be searched : "))
    print()
    
    # in bookRoom method we stored details in terms of room number
    if roomNo in h.searchRooms.keys():
        print(f"Room {roomNo} has been booked to {h.searchRooms[roomNo]}")
        print("--------------------------------------------------------")
        # gets the customer name  
        name = h.searchRooms[roomNo]
        print(f"Name : {name} \nContact No. : {data[name][0]} \nNo. of rooms : {data[name][1]} \nDuration of Staying : {data[name][2]} \n")
        print("----------------------------------------------------")
        print()
        
    # condition when room number not exists in the hotel    
    elif(roomNo > 100 ):   
        print("----------------------------------------------------")
        print("xxxx Invalid Room Number xxxx")
        print("----------------------------------------------------")
        print()
    else:
        print("----------------------------------------------------")
        print("***** Room has not been booked *****")
        print("----------------------------------------------------")
        print()
        
```

```python
# checks the availability of room which customer asked 
def checkAvailability():
    checkRoom=int(input("Enter the room number to be checked for availability : "))
    
    # simply check given room number in there in availRoom datalist
    if checkRoom in h.availRooms:
        print("----------------------------------------------------")
        print(f"==== Room {checkRoom} is available ====")
        print("----------------------------------------------------")
        print()
        
    # if given room is present in reserved room datalist
    elif checkRoom in h.reservedRooms:
        print("----------------------------------------------------")
        print(f"==== Room {checkRoom} is unavailable ===")
        print("----------------------------------------------------")
        print()
        
    else:
        print("----------------------------------------------------")
        print("xxxx Invalid Room Number xxxx")
        print("----------------------------------------------------")
        print()
        return
```

### Editing A Record

```python
# edit booking record
def edit():
    roomNo = int(input("Enter the room number to edit booking record :: "))
    
    if roomNo in h.searchRooms.keys():
        print(f"Old Record of Room No. {roomNo} ")
        print("--------------------------------------------------------")
        # gets the customer old name  
        oldname = h.searchRooms[roomNo]
        print(f"Name : {oldname} \nContact No. : {data[oldname][0]} \nNo. of rooms : {data[oldname][1]} \nDuration of Staying : {data[oldname][2]} \n")
        print()
        print(f"Enter New Record ")
        name = input("Customer Name : ")
        phNum = input("Contact No. : ")
        stayDuration = int(input("Duration stayed (days) in room : "))
        
        # new contact no and stay duration will update in old record
        data[oldname][0] = phNum
        data[oldname][2] = stayDuration
        # here old name record will pop out in new name 
        data[name] = data.pop(oldname)
        # updating details in searchRooms datalist
        h.searchRooms[roomNo] = name
        
        print("----------------------------------------------------")
        print("Preview Change Record :: ")
        print(f"Name : {name} \nContact No. : {data[name][0]} \nNo. of rooms : {data[name][1]} \nDuration of Staying : {data[name][2]} \n")
        print("----------------------------------------------------")
        print()
        
    elif(roomNo > 100 ):      
        print("----------------------------------------------------")
        print("xxxx Invalid Room Number xxxx")
        print("----------------------------------------------------")
        print()
    else:
        print("----------------------------------------------------")
        print("***** Only Booked Room Can Be Edited *****")
        print("----------------------------------------------------")
        print()
```

### Checkout From Hotel (delete)

```python
# deleting a record or Checkout from hotel
def checkout():
    print("************** Checkout **************")
    name = input("Enter the Customer Name who wants to Checkout :: ")
    
    if name in data.keys():
        # get total no rooms customer boked
        bookedRooms = data[name][3]
        # checked out rooms(empty rooms) will add up in available rooms
        h.availRooms.extend(bookedRooms)
        # will empty the seacrhRooms and reservedRooms datalist 
        for roomNo in bookedRooms:
            h.searchRooms.pop(roomNo)
            h.reservedRooms.remove(roomNo)
        # delete the record from database of checked out person
        del data[name]
        print()
        print("---------- Successfully Checked Out ----------")
        
    else:
        print("----------------------------------------------------")
        print("xxxx Record Not Found xxxx")
        print("----------------------------------------------------")
        print()
    
```

### Home Menu

```python
# its the starting home menu options
def start():
    while True:
        print("*** Home Menu ***")
        print(" 1. Booking")
        print(" 2. Enquiry")
        print(" 3. Edit Record")
        print(" 4. Checkout")
        print(" 5. Exit")
        print()
        
        choice = int(input("Enter your choice :: "))
        print()
        
        if choice == 1:
            booking()
            print()
        elif choice == 2:
            enquiry()
            print()
        elif choice == 3:
            edit()
            print()
        elif choice == 4:
            checkout()
            print()
        elif choice == 5:
            print("Thank You for using Hotel Management System. Visit Again :) ")
            print("X"*70)
            print()
            break
        else:
            print("xxxxx Please enter valid number, 1/2/3/4/5 xxxxx")
            print()
            
```
