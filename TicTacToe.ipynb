```python
#play game function 
def play_game():
    # # Global variables
       global board, game_not_exit, winner, current_player, quit

       board = ["-", "-", "-",
                "-", "-", "-",
                "-", "-", "-"]

       # Lets us know if the game is over
       game_not_exit = True

       # Who is the current winner
       winner = None

       # Current player i.e 'X' goes first
       current_player = "X";

       # Initially quit flag is turned off
       quit = False
       #calls for selection of mode 
       select_mode()
```

```python
#display board
def display_board():
       print(board[0]+" | "+board[1]+" | "+board[2])
       print(board[3]+" | "+board[4]+" | "+board[5])
       print(board[6]+" | "+board[7]+" | "+board[8])
```

```python
#select mode function between manual and auto
def select_mode():
       print("Select playing mode : ")
       print("1. Manual Mode")
       print("2. Auto Mode")
        
       choice=input("Choice :: ")

       if choice == "1" :
              manual_mode()
       else :
              auto_mode()
```

## Manual Mode

```python
# manual mode function
def manual_mode():
       print("<=============> Manual Mode <===============>")

       global player_data
       # input player name
       player1_name = input("Enter player name who wants 'X' : ")
       player2_name = input("Enter player name who wants 'O' : ")

       # displaying player name who choose X and O     
       print(player1_name + " --> Choose 'X'")
       print(player2_name + " --> Choose 'O'")

       #display initial empty tic tac toe board
       print()
       display_board()
       print()

       player_data = {
              'X': player1_name,
              'O': player2_name
       }
        
       # Until game is not over or unless anyone wins or quit 
       while game_not_exit:
              handle_turn(current_player)
              check_game_over()
              change_player_turn()

       #The game has ended and who won 
       if winner == "X" or winner == "O":
              print("----------------------------------")
              print("|  "+player_data[winner]+ " WON  |")
              print("----------------------------------")
        
    # checks for quiting of game in between 
       elif quit == True:
              won = player_data[current_player]
              change_player_turn()
              lose = player_data[current_player]
              print("------------------------------------------------------")
              print("|  \""+lose+"\" has QUIT the match, \""+won+"\" WON  |")
              print("------------------------------------------------------")
      
    #The game has ended but tied
       elif winner == None :
              print("################")
              print("#  Match TIED  #")
              print("################")

       print()
    #If players decides to play again or not
       play_again()
```

```python
# handle turn function handles valid p0sition input of the player
# here, parameter player is players mark
def handle_turn(player):

       #get position from player
       print(player_data[player]+"'s turn.")
       position = input("Choose a position from 1-9 or quit : ")

       # If any player enters "quit" in between the game ends 
       if position == "quit":
            #calls the quit in between to end the game
            quit_in_between()
            return
        
       #check whether the input position is valid or not
       valid=False
       while not valid:
              #check only valid position is inputed
              while position not in ["1","2","3","4","5","6","7","8","9"]:
                     print("Enter value only between 1-9 or quit . Go Again")
                     position = input("Choose a position from 1-9 or quit : ")

             #position will returned in string convert it to int and 
             # -1 the to match the actual index            
              position = int(position)-1

              # entered value not be repeated
              if board[position] == "-":
                     valid=True
              else:
                     print("Already value exists. Go Again")

       board[position] = player
    # Display current status of the game board
       display_board()
```

```python
# check game over either of player win or tied match
def check_game_over():
       check_win()
       check_tie()
```

```python
# check for winner from row , column or diagonals
# if 3 consecutive in straight line same mark appears in row
# column or digonal player wins

def check_win():
       #set up winnner as global variable
       global winner

       #get who is the row or column or diagonal winner
       row_winner=check_row()
       column_winner=check_column()
       diagonal_winner=check_diagonals()

       if row_winner:
              winner=row_winner
       elif column_winner:
              winner=column_winner
       elif diagonal_winner:
              winner=diagonal_winner
       else:
              winner=None

```

```python
# check winner in row category
def check_row():
       #set game_not_exit as global variable
       global game_not_exit
       # check any of the row have same value but also not equal to "-"
       row1=board[0]==board[1]==board[2] != "-"
       row2=board[3]==board[4]==board[5] != "-"
       row3=board[6]==board[7]==board[8] != "-"

       #if any of the row gets same game will exit
       if row1 or row2 or row3:
              game_not_exit =False
       #return winner
       if row1:
              return board[0]
       elif row2:
              return board[3]
       elif row3:
              return board[6]
       else:
              return None
```

```python
# check winner in column category
def check_column():
       # set game_not_exit as global variable
       global game_not_exit
       # check any of the column have same value but also not equal to "-"
       column1 = board[0] == board[3] == board[6] != "-"
       column2 = board[1] == board[4] == board[7] != "-"
       column3 = board[2] == board[5] == board[8] != "-"

       # if any of the column gets same game will exit
       if column1 or column2 or column3:
              game_not_exit = False

       # return the winner
       if column1:
              return board[0]
       elif column2:
              return board[1]
       elif column3:
              return board[2]
       else:
              return None
```

```python
# check winner in diagonal category
def check_diagonals():
       # set game_not_exit as global variable
       global game_not_exit
       # check any of the diagonal have same value but also not equal to "-"
       diagonal1 = board[0] == board[4] == board[8] != "-"
       diagonal2 = board[2] == board[4] == board[6] != "-"

       # if any of the diaganal gets same game will exit
       if diagonal1 or diagonal2:
              game_not_exit = False

       # return the winner
       if diagonal1:
              return board[0]
       elif diagonal2:
              return board[2]

       else:
              return None
```

```python
# check if all the position in the board is filled the match becomes tie
def check_tie():
       #set game_not_exit as global variable
       global game_not_exit
    
       #if board is full
       if "-" not in board:
              game_not_exit = False
              return True
       else:
              return False
```

```python
# change the player turns alternatively
def change_player_turn():
       #set current_player as global variable
       global current_player
    
       #if current player is X chnage it to O and vice versa
       if current_player == "X":
              current_player="O"
       elif current_player == "O":
              current_player="X"
       return
```

```python
# Bot gets the random index which are not visited yet
def random_place():
    #selection list will contain non visited index position
       selection = possible_place()
    # current location gets the random computer choosed index and returns    
       current_location = random.choice(selection)
       return current_location
```

```python
## quit in between
def quit_in_between():
    # set global variable 
       global game_not_exit, quit

     # force the game to exit at current position   
       game_not_exit = False
     # turn on the quit flag to quit the game   
       quit = True
       return
```
